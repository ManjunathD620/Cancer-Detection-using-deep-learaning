```python
import pandas as pd

data = pd.read_csv("/content/data.csv")
data.head()
```

```python
data = data.replace({"diagnosis": {'M': 0, 'B': 1}})
data = data.iloc[:, :-20]
x = data.drop("diagnosis" ,axis=1).values
y = data["diagnosis"].values
print(x.shape)
print(y.shape)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=10)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(1000, input_dim=12, activation="relu"))
model.add(Dense(500, activation="relu"))
model.add(Dense(300, activation="relu"))
model.add(Dense(100, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(x_train,y_train, epochs=20, batch_size=10,validation_data=(x_test,y_test))
```

```python
import random
import numpy as np

rand = random.randint(0,x_test.shape[0])
subject = x_test[rand]
subject = np.expand_dims(subject,axis=0)
prediction = model.predict(subject)
print("Predicted = "+str(round(prediction[0][0]))+"\n"+"actual = "+str(y_test[rand]))
```
